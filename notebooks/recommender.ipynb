```python
import pandas as pd

df = pd.read_csv("../data/cleaned_ratings.csv")

df.head()
```

```python
popular = df.groupby("product_id")["rating"].mean()
popular = popular.sort_values(ascending=False)

popular.head(10)
```

```python
count = df.groupby("product_id")["rating"].count()

popular_df = pd.DataFrame({
    "avg_rating": popular,
    "num_reviews": count
})

popular_df = popular_df.sort_values(
    by=["avg_rating", "num_reviews"],
    ascending=False
)

popular_df.head(10)
```

```python
user_counts = df["user_id"].value_counts()

active_users = user_counts[user_counts > 20].index

df_small = df[df["user_id"].isin(active_users)]

df_small.shape
```

```python
product_counts = df_small["product_id"].value_counts()

popular_products = product_counts[product_counts > 20].index

df_small = df_small[df_small["product_id"].isin(popular_products)]

df_small.shape
```

```python
user_item = df_small.pivot_table(
    index="user_id",
    columns="product_id",
    values="rating"
)
```

```python
user_item.shape
```

```python
user_item_filled = user_item.fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(user_item_filled)
```
