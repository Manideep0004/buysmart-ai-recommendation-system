```python
import pandas as pd

df = pd.read_csv(
    "../data/ratings.csv",
    header=None,
    names=["product_id", "user_id", "rating", "timestamp"]
)

df.head()
df.columns
```

```python
df.columns
```

```python
df = df[["user_id", "product_id", "rating"]]
```

reduce dataset

```python
user_counts = df["user_id"].value_counts()
product_counts = df["product_id"].value_counts()

df = df[df["user_id"].isin(user_counts[user_counts > 10].index)]
df = df[df["product_id"].isin(product_counts[product_counts > 10].index)]

df.shape
```

sparse matrix

```python
df["user_id"] = df["user_id"].astype("category")
df["product_id"] = df["product_id"].astype("category")

df["user_index"] = df["user_id"].cat.codes
df["product_index"] = df["product_id"].cat.codes

from scipy.sparse import csr_matrix

matrix = csr_matrix(
    (df["rating"],
     (df["user_index"], df["product_index"]))
)

matrix
```
